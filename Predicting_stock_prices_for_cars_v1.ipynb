```python
# ===== ФУНКЦИИ с ПРОШЛЫХ ПРОЕКТОВ =====

def plot_learning_curve (title, model, features, target, score, train_sizes): 
    '''Использует встроенную функцию learning_curve
    и рисует кривые валидации на отдельном графике'''
    
    train_sizes,train_scores, validation_scores = learning_curve(
                                                estimator =  model,
                                                X = features,                                      
                                                y = target,        
                                                scoring = score,
                                                cv=2,
                                                train_sizes = train_sizes, shuffle=True)                                        
                                              
        
    # метрики качества модели, которые мы получили для обучающих датасетов cv=5
    # вычисляем среднее арифметическое для графика кривой
    train_scores_mean = train_scores.mean(axis = 1)
    validation_scores_mean = validation_scores.mean(axis = 1)

    plt.style.use('seaborn')
    plt.plot(train_sizes, train_scores_mean, label = 'Training error')
    plt.plot(train_sizes, validation_scores_mean, label = 'Validation error')
    plt.ylabel(score, fontsize = 14)
    plt.xlabel('Training set size', fontsize = 14)
    plt.title(title, fontsize = 18, y = 1.03)
    plt.legend()
    plt.ylim(-1600, -300)
    
    plt.show()

# ===== ФУНКЦИИ с ТЕКУЩЕГО ПРОЕКТА =====

def format_date(data, columns_list):
    ''' меняет тип данных на datetime с нулевым временем
    в указанных столбцах columns_list
    '''
    d = data.copy()
    for column in columns_list:
        # используем normalize чтобы оставить у всех нулевое время
        # в этом случае будет отображаться только дата
        d[column] = pd.to_datetime(d[column]).dt.normalize()
        
    return d    


def get_missing_values(data):
    ''' Вычисляет количество и долю пропусков 
    и возвращает сгенерированную таблицу
    '''    
    d = (data
         .isna().sum().rename('missing_count').to_frame()
         .join(df.isna().mean().rename('missing_ratio').to_frame())
         .apply(lambda x: x[x > 0])
         .sort_values(by='missing_count', ascending=False)
        )
   
    return(d)

def fill_nan_unknown(data, column_list):
    ''' Заполняет пропуски строкой "unknown"
    в указанных столбцах column_list
    '''    
    d = data.copy()    
    d[column_list] = d[column_list].fillna('unknown')
    
    return d

def train_and_get_results(model, features, target, cv_folds): 
    ''' 
    Обучает модель/пайплайн и выводит RMSE-ошибки на кросс-валидации
    '''

    rmse_errors = cross_val_score(model, features, target, scoring = 'neg_root_mean_squared_error', cv=cv_folds)      
    rmse_error = rmse_errors.mean()*(-1)    
     
    print(f"RMSE error (mean): {rmse_error: .2f} | {rmse_errors * (-1)}")  
    print("=======")
    
    
def testing_pipeline_crossval(model, encoding_pipeline, features, target, cv_folds): 
    ''' 
    Cоздает пайплайн (модель + кодирование)
    выводит RMSE-ошибки и R2-score на кросс-валидации
    '''
    pipeline = Pipeline(steps=[   
                              ('encoder', encoding_pipeline),
                              ('model', model)
                        ])
    
    rmse_errors = cross_val_score(pipeline, features, target, scoring = 'neg_root_mean_squared_error', cv=cv_folds).round(2)      
    rmse_error = rmse_errors.mean()*(-1) 
    
    r2_scores = cross_val_score(pipeline, features, target, scoring = 'r2', cv=cv_folds).round(3)
    r2_score = r2_scores.mean()
    
    print(f"RMSE error: {rmse_error: .2f} | {rmse_errors * (-1)}")  
    print(f"R2 score:   {r2_score: .3f}   | {r2_scores}")  
        
    print("=======")
```

```python
# Сохраняем базовую модель для дальнейших экспериментов
rf_base_model = RandomForestRegressor(max_depth = 20, n_estimators = 20, random_state=1234)
```

####  Влияние кодирования на результат (создаем пайплайны)

Создадим `несколько пайплайнов` для разных вариантов **кодирования категорий**

```python
# еще раз создадим encoder для OrdinalEncoding
ordinal_encoder = ce.OrdinalEncoder(cols=train_df.select_dtypes(include=[object]).columns)

# Переводим в ONE-HOT кодирование 'not_repaired', 'vehicle_type', 'auto_fuel_type'
ohe_encode_pipeline  = Pipeline(steps=[   
                       ('encoder_ordinal',ce.OrdinalEncoder(cols=['auto_model', 'auto_brand', 'auto_gearbox'], return_df=True)),
                       ('encoder_onehot', ce.OneHotEncoder(cols=['not_repaired', 'vehicle_type', 'auto_fuel_type'], use_cat_names=True, return_df=True))                                
                        ])

# Вместо ONE_HOT используем DUMMY кодирование для категорий
dummy_column_pipeline = ColumnTransformer(
                            transformers=[  

                            # вместо encoder можно добавить целый пайплайн (например, по препроцессингу)
                            ('encoder_ordinal', ce.OrdinalEncoder(), ['auto_model', 'auto_brand', 'auto_gearbox']),

                            # с параметром drop='first' мы получим dummy-encoder
                            ('encoder_onehot', OneHotEncoder(drop='first'), ['not_repaired', 'vehicle_type', 'auto_fuel_type'])                                
                            ], 
                        ## (!) очень важно использовать параметр passthrough
                        ## иначе будут откинуты не перечисленные параметры датасета 
                        remainder='passthrough')

# Вместо Ordinal Encoding используем Target Encoding (для 'auto_model' и 'auto_brand')
target_encode_pipeline  = Pipeline(steps=[   
                       ('encoder_target', ce.TargetEncoder(cols=['auto_model', 'auto_brand'], smoothing=0,return_df=True)),
                       ('encoder_ordinal',ce.OrdinalEncoder(cols=['auto_gearbox'], return_df=True)),
                       ('encoder_onehot', ce.OneHotEncoder(cols=['not_repaired', 'vehicle_type', 'auto_fuel_type'], use_cat_names=True, return_df=True))                               
                        ])
```

- Наш стартовый результат с OrdinalEncoding -  `RMSE = 1593` (и RMSE = 1573 для 100 деревьев)


- **Исправим кодирование** и проверим, повезет ли нам с качеством предсказаний)

```python
%%time
print(f"Добавляем ONE-HOT кодирование для категорий\n")
print(rf_base_model)
train_and_get_results(make_pipeline(ohe_encode_pipeline, rf_base_model), 
                      train_df.drop(['price'], axis=1), train_df.price, cv_folds=3)
```

Отлично! Протестируем вариант с `dummy-encoder` (удалим  лишний столбец в OHE). Станут ли результаты лучше?

```python
%%time
print("Вместо ONE_HOT используем DUMMY кодирование для категорий\n")
train_and_get_results(make_pipeline(dummy_column_pipeline, rf_base_model), 
                      train_df.drop(['price'], axis=1), train_df.price, cv_folds=3)
```

Интересно, похоже, **`ColumnTransformer`** работает **быстрее**, чем варианты с **`Pipeline`**. Протестировано несколько раз (в т.ч. с такими же encoders, как и в `ohe_encode_pipeline`)

```python
%%time
print("Добавляем TARGET кодирование \n")
print(rf_base_model)
train_and_get_results(make_pipeline(target_encode_pipeline, rf_base_model), 
                      train_df.drop(['price'], axis=1), train_df.price, cv_folds=3)
```
